#### (4) One way to deal with noisy data is to use a moving average. Plot the moving average of the data for two different windows. How does this help visualize the data? What is lost using this approach?

```python
import pandas as pd

plt.figure(6, figsize=(12.5, 7.5), dpi=100)

plt.plot(X, dataY, '.', color="#8795E8", alpha=0.25)
plt.plot(X, pd.rolling_mean(pd.rolling_mean(dataY, round((len(dataY)*(1/(24*2))))), round((len(dataY)*(1/(24*2))))), c="#661246", alpha=0.75, label='moving average of the moving average of heart rate every 30 minutes') # every 30 minutes
plt.xticks(np.arange(min(X), max(X)+1, 1.0))
plt.yticks(np.arange(min(dataY), max(dataY)+5, 5))
plt.grid(True, alpha=0.25)
plt.xlim(np.min(X), np.max(X))
plt.ylabel('heart rate (beats per minute)')
plt.xlabel('time (hours)')
plt.legend()

plt.show()
```

```python
plt.figure(6, figsize=(12.5, 7.5), dpi=100)

plt.plot(X, dataY, '.', color="#8795E8", alpha=0.25)
plt.plot(X, pd.rolling_mean(dataY, round((len(dataY)*(1/(24*12))))), c="#f9897b", alpha=0.85, label='moving average of the heart rate every 5 minutes') # every 5 minutes
plt.xticks(np.arange(min(X), max(X)+1, 1.0))
plt.yticks(np.arange(min(dataY), max(dataY)+5, 5))
plt.grid(True, alpha=0.25)
plt.xlim(np.min(X), np.max(X))
plt.ylabel('heart rate (beats per minute)')
plt.xlabel('time (hours)')
plt.legend()

plt.show()
```

Taking a moving average smooths the data and removes some noise. We see however that there is some amount of data lost depending on the size of the window. If the window one uses is as large as the spacing between data points, then no data is lost. However, taking larger and larger windows requires that the average only shows up at the first time the window is completed. In my first chart, the first average is only plotted after the first half hour. In my second chart, the first average only shows up at the end of the first 5 minutes. It does prevent jump discontinuities between data as well and makes it a continuous differentiable function (after successive repeated moving averages). In the first chart, I took the moving average twice, and it is remarkably smoother than the original data.

#### (5) We're going to fit an HMM model to this data with three underlying states. This involves three sets of parameters about the modeldescribe each briefly.

__Initial Probabilities:__

Probabilites that describe the chance of being in a certain state initially. All probabilities in this group sum to one. 

$$S_{t}\in \{X, Y, Z\}$$

-

where $t$ is the state count ($t=0$ is the initial step) and $X,Y,Z$ are the three underlying hidden states.

$$p(S_{t=0}=X)+p(S_{t=0}=y)+p(S_{t=0}=Z)=1$$

-

__Transmission Probabilities:__

Probabilites that describe the chance of transitioning into a different state (to $S_{t+1}$) from a certain state (from $S_t$). There will be 9 probabilities here, one for each given state of the three states transitionining into one of the same three states. Each of the probabilities transitioning to one of the three states from one given state sum to one.

$$p(S_{t+1}=X|S_{t}=X)+p(S_{t+1}=Y|S_{t}=X)+p(S_{t+1}=Z|S_{t}=X)=1$$

-

$$p(S_{t+1}=X|S_{t}=Y)+p(S_{t+1}=Y|S_{t}=Y)+p(S_{t+1}=Z|S_{t}=Y)=1$$

-

$$p(S_{t+1}=X|S_{t}=Z)+p(S_{t+1}=Y|S_{t}=Z)+p(S_{t+1}=Z|S_{t}=Z)=1$$

-

It is clear to see that these probabilities can be placed into a square matrix called the transition matrix.

__Emission Probabilities:__

Probabilites that describe the chance of obtaining a certain outcome in the observable range of data $h$ given a certain state. All probabilities of each outcome $h_i$ for a given state sum to one.

$$\sum_{\forall y}(H_{t}=h_i|S_{t}=X)=1$$

-

$$\sum_{\forall y}(H_{t}=h_i|S_{t}=Y)=1$$

-

$$\sum_{\forall y}(H_{t}=h_i|S_{t}=Z)=1$$

-

#### (6) Fit an HMM model with Gaussian emissions to the data. Use any transformations as necessary.

How does the model respond to rapid state changes? Overlay the raw measurements and HMM state over time.

```python
from hmmlearn.hmm import GaussianHMM

dataY_HMM_fit_input = np.reshape(dataY, newshape=(len(dataY),1))

model = GaussianHMM(n_components=3, covariance_type='full').fit(dataY_HMM_fit_input)
hidden_states = model.predict(dataY_HMM_fit_input)
model_decode_state = model.decode(dataY_HMM_fit_input)[1]
model_decode_state_scaled = [(np.squeeze(model.means_[y])) for y in model_decode_state]
model_decode_state_scaled_plus_std =  [(np.squeeze(model.means_[y]))+(np.squeeze(np.diag(model.covars_[y]))**(0.5)) for y in model_decode_state]
model_decode_state_scaled_minus_std =  [(np.squeeze(model.means_[y]))-(np.squeeze(np.diag(model.covars_[y]))**(0.5)) for y in model_decode_state]
```

```python
print("Transition matrix:")
print(model.transmat_)

print("\nMeans and vars of each hidden state:")
for i in range(model.n_components):
    print("\nhidden state {}".format(i+1))
    print("mean =    ", (np.squeeze(model.means_[i])))
    print("std dev = ", (np.squeeze(np.diag(model.covars_[i]))**(0.5)))
```

```python
plt.figure(7, figsize=(12.5, 7.5), dpi=100)

plt.plot(X, dataY, '.', color="#8795E8", alpha=0.25, label='observed data')
plt.plot(X, model_decode_state_scaled, '.', color="#f9247e", alpha=0.75, label='state means')
plt.fill_between(X,model_decode_state_scaled_minus_std,model_decode_state_scaled_plus_std, alpha=0.3, color="#d7509f", label="1 std dev above and below the state means")

plt.xticks(np.arange(min(X), max(X)+1, 1.0))
plt.yticks(np.arange(min(dataY), max(dataY)+5, 5))
plt.grid(True, alpha=0.25)
plt.xlim(np.min(X), np.max(X))
plt.legend()
plt.ylabel('heart rate (beats per minute)')
plt.xlabel('time (hours)')

plt.show()
```

When the heart rate begins to transition into different regions the model jumps discontinuously since the states are discrete and not continuous. It does recognize that data prior to and after a rapid state change must belong to different states.

#### (7) From the HMM model determine, with a confidence interval, when the patient woke up in the morning. Compare this prediction to the data you obtain with the moving average.

Hint: `model.predict_proba(X)` can give you the probability of being in each state at each time point.

```python
a, b, c = zip(*model.predict_proba(dataY_HMM_fit_input))

if model.means_[0]< model.means_[1] and model.means_[0]< model.means_[2]:
    sleeping = a
    if model.means_[1]< model.means_[2]:
        woke = b
        strenuous = c
    else:
        woke = c
        strenuous = b
elif model.means_[1]< model.means_[0] and model.means_[1]< model.means_[2]:
    sleeping = b
    if model.means_[0]< model.means_[2]:
        woke = a
        strenuous = c
    else:
        woke = c
        strenuous = a
else:
    sleeping = c
    if model.means_[0]< model.means_[1]:
        woke = a
        strenuous = b
    else:
        woke = b
        strenuous = c
        
sleeping_not_data = [x + y for x, y in zip(woke, strenuous)]

plt.figure(9, figsize=(12.5, 7.5), dpi=100)
plt.plot(X,woke, color="#392682")
plt.fill_between(X,0,woke, alpha=0.25,color="#7a3a9a", label="Awake")
plt.plot(X,sleeping,c="#28ada8")
plt.fill_between(X,0,sleeping,alpha=0.25,color="#83dde0",label="Sleeping")
plt.plot(X,strenuous,c="#569874")
plt.fill_between(X,0,strenuous,alpha=0.25,color="#296656", label="Strenous")
plt.xlim(6, 7)
plt.xticks(np.arange(6, 7+(5/60), 5/60))
plt.ylim(0, 1)
plt.legend()
plt.ylabel('probability')
plt.xlabel('time (hours)')
plt.show()
```

```python
plt.figure(9, figsize=(12.5, 7.5), dpi=100)
plt.plot(X,sleeping_not_data, color="#392682")
plt.fill_between(X,0,sleeping_not_data, alpha=0.25,color="#ffde8b", label="Awake or Strenuous")
plt.plot(X,sleeping,c="#28ada8", label="Sleeping (not awake or strenuous)")
plt.legend(loc='upper right')
plt.ylabel('probability')
plt.xlabel('time (hours)')

plt.xlim(6.1667, 6.3333)
plt.xticks(np.arange(6.1667, 6.3333+(5/60), 5/60))
plt.ylim(0, 1)
plt.show()
```

```python
plt.figure(9, figsize=(12.5, 7.5), dpi=100)
plt.plot(X,sleeping_not_data, color="#392682")
plt.fill_between(X,0,sleeping_not_data, alpha=0.25,color="#ffde8b", label="Awake or Strenuous")
plt.plot(X,sleeping,c="#28ada8", label='Sleeping (not awake or strenuous)')

plt.xlim(6.58333, 6.75)
plt.xticks(np.arange(6.58333, 6.6666667+(5/60), 5/60))
plt.ylabel('probability')
plt.xlabel('time (hours)')
plt.ylim(0, 1)
plt.legend(loc='upper right')
plt.show()
```

It seems like this person woke up for the day around 0640 which matches our moving average pretty well. He may have also snoozed his alarm at 0630 according to our first histogram.
