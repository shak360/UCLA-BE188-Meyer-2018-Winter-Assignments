```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as pt
import sklearn
from sklearn.datasets import load_boston
%matplotlib inline
```

#### Load in the Boston dataset. Get an overview of the data.

```python
boston = load_boston() # dictionary

print(boston.keys())
```

```python
bostonDF = pd.DataFrame(boston.data)
bostonDF.head()
```

```python
print(bostonDF.shape)
```

```python
bostonDF.columns = boston.feature_names
```

```python
bostonDF.head()
```

```python
bostonDF['PRICE'] = boston.target
bostonDF.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression(n_jobs=(-1), )

X = bostonDF.drop(['PRICE'], axis=1)
X.head()
Y = bostonDF.PRICE
```

```python
linreg.fit(X, Y)
```

```python
param_results = pd.DataFrame(list(zip(X.columns, linreg.coef_)))
param_results.columns = ['Features', 'Estimated Coeffs']
```
